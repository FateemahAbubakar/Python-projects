```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv_imshow
```

```python
# Load the image (OpenCV loads as BGR)
img_bgr = cv2.imread("Corrupted Student Picture.png")
```

```python
# Convert to RGB for proper display in matplotlib
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
alpha = 1.5   # contrast
beta = 40     # brightness
adjusted = cv2.convertScaleAbs(img_rgb, alpha=alpha, beta=beta)
```

```python
r, g, b = cv2.split(adjusted)
```

```python
restored = cv2.merge([b, g, r])
```

```python
# Show images
plt.figure(figsize=(15,6))
plt.subplot(1,3,1)
plt.title("Corrupted Image")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Brightness/Contrast Adjusted")
plt.imshow(adjusted)
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Restored Image")
plt.imshow(restored)
plt.axis("off")

plt.show()
```
